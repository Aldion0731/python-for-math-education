```python
import sympy
import numpy as np
from sympy import *
import pandas as pd
import re
x, y, z = symbols("x y z")
from itertools import combinations_with_replacement
```

```python
def generate_factorization_questions(a):
    a = list(np.random.randint(-a, a +1, size=1)) #generates a random coefficient of x^2 between the given a values
    c = list(np.random.randint(-20, 21, size=1)) #generates a random constant term between the given a values
    if a[0] == 0:
        a = np.random.randint(1, 5, size=1)[0] #This ensures a != 0
    else:
        a = a[0]
    
    if c[0] == 0:
        c = np.random.randint(1, 21, size=1)[0] #This ensures c != 0
    else:
        c = c[0]
  
  
    
    ac = a * c #We need the b term to be the sum of two numbers whose product is a * c. So, we must know a * c
    if ac > 0:
        ac_range = range(-ac, ac + 1)
    else:
        ac_range = range(ac, -ac + 1) #This range will contain all the possible values of b1 and b2
   
    b_list = [(b1, b2) for (b1, b2) in combinations_with_replacement(ac_range, 2) if b1 * b2 == ac]
    #This will generate all the possible pairs of values which could sum to ac
    
    num_possibilities = len(b_list) #These are the number of possible pairs
       
    choice_index = np.random.randint(0, num_possibilities) #We will use this to select exactly one pair
    
    b_pair = b_list[choice_index]
    b = sum(b_pair) #This will be the b term of the quadratic s.t. our quadratic will certainly have 2 bino
    
    
    
    
    sym_eq = a * x ** 2 + b * x + c #Generates a sympy equation which the .factor() function can easily factorize.
    
         
    if a == 1:
        a = ""
    if a == -1:
        a = '-'
    if b == 1:
        b = ""
    if b == -1:
        b = '-'
        

    if b == 0:
         eq = '{}x\u00b2 + {}'.format(a,  c)
    else:
        eq = '{}x\u00b2 + {}x + {}'.format(a, b, c)
        
           
    eq = re.sub(r'[+] [-]', '-', eq)
    
    
    return eq, sym_eq.factor()

    
    
```

```python
generate_factorization_questions(6)
```

```python
def generate_spreadheet(num_questions, sheet_name, max_a=6):
    questions = []
    answers = []
    for i in range(num_questions):
        q, a = generate_factorization_questions(max_a)
        questions.append(q)
        answers.append(a)
        database = {'questions': questions, 'answers': answers}
        df = pd.DataFrame(database)
        df['answers'] = df['answers'].apply(str)
        df['answers'] = df['answers'].replace(')**2', ')\u00b2')
        df['answers'] = df['answers'].apply(lambda x: re.sub(r'[*]', '', x))
        df.to_excel(sheet_name)
        
    return df
```

```python
generate_spreadheet(1000, '1000 New Factorization Questions.xlsx', 7)
```
